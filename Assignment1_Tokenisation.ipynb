```python
import re
```

```python
with open("Tokenization_Material.html", "r", encoding = "utf-8") as f:
    raw_text = f.read()
```

```python
text = re.split(r'(\s)', raw_text)
# print(text)
```

```python
text = re.split(r'([,.]|\s)', raw_text)
# print(text)
```

```python
text = re.split(r'([,.!><@#$%^&*()+=?]|--|\s)', raw_text)
# print(text)
```

```python
text = [item.strip() for item in text if item.strip()]
print(text)
```
