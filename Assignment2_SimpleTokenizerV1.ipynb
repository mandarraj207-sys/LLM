```python
import re
```

```python
with open("SimpleTokenizerV1_Material.html", "r", encoding = "utf-8") as f:
    raw_text = f.read()
```

```python
text = re.split(r'([,.:;?_!()@#$%^&*<>\']|--|\s)', raw_text)
text = [item.strip() for item in text if item.strip()]
```

```python
all_words = sorted(set(text))
vocab_size = len(all_words)
```

```python
vocab = {token : integer for integer, token in enumerate(all_words)}
```

```python
print(f"vocab size is {vocab_size} \n")
print(vocab)
```
