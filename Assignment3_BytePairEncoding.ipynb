```python
import re

from importlib.metadata import version
import tiktoken
print("tiktoken version:", version("tiktoken"))
```

```python
class Build_Vocab:
    with open("vocabForLLM.txt", "r", encoding = "utf-8") as f:
        raw_text = f.read()

    text = re.split(r'([,.<>?!@#$%^&*()+=\'\"]|--|\s)', raw_text)
    text = [item.strip() for item in text if item.strip()]

    all_words = sorted(set(text))
    vocab_size = len(all_words)
    
    all_words.insert(0, "<UNK>")

    vocab = {token : integer for integer, token in enumerate(all_words)}
```

```python
class Encode_Text:
    with open ("Validate_data.txt", "r", encoding = "utf-8") as f:
        input_text = f.read()

    enc = tiktoken.get_encoding("gpt2")
    encoded_tokens = enc.encode(input_text)
    decoded_text = enc.decode(encoded_tokens)

    print("Input Text:")
    print(input_text)

    print("\nEncoded Tokens:")
    print(encoded_tokens)

    print("\nDecoded Text:")
    print(decoded_text)
```
